```python
data_folder = '/Users/ryznerf/Documents/0_MIT/Spring_2022/0_Projects/3_Networks_project/community-detetction-deep-learning/Data/'
files = ['toy_dataset.pkl', 'complex_1.pkl', 'complex_2.pkl']
```

```python
# CASE 1
g = Utils.unpickle_object(data_folder + files[0])
true, pred = Non_DL_Methods.Louvain_algorithm(g)
NMI = normalized_mutual_info_score(np.array(list(true.values())), np.array(list(pred.values())))
Plotting.graph_cluster_plot(g=g, partition=pred, title='Predicted - Louvain case 1 - NMI: ' + str(np.round(NMI, 3)))
```
