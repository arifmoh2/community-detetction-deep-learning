```python
from matplotlib.cm import get_cmap
import networkx as nx
import numpy as np
import os
import pickle
import matplotlib.pyplot as plt

#For saving results
data_dir = os.path.join((os.path.dirname(os.getcwd())), 'Data')

#For plotting
name = "Accent"
cmap = get_cmap(name)
colors = cmap.colors
```

## Task 1: Make a toy dataset that can be used to test implementations of our community detection algorithms

Networkx gives an implementation of the stochastic block model and alos tells us which community each ode belongs to. As a visual illustration, see below:

```python
p = 0.5    #probability of edges within a cluster
q = 0.01 #probability of edges across clusters

sizes = [10, 10, 25]
n = len(sizes)

probs = np.eye(n)*p + (1 - np.eye(n))*q

g = nx.stochastic_block_model(sizes, probs, seed=0)

nx.draw(g,node_color=[colors[i] for i, nodes in enumerate(g.graph['partition']) for node in nodes], with_labels=False)
```

Let's work with this 3 community toy example for the time being. Later we can play with sizes and edge probabilities.

```python
file_path = os.path.join(data_dir, 'toy_dataset.pkl')
with open(file_path, 'wb') as f:
    pickle.dump(g, f)
```
